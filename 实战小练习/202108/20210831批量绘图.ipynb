```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def drawfig(df,list1,label1,figname1,title1):
    plt.figure(figsize=(25,10))
    plt.rcParams['font.sans-serif'] = ['SimHei']

    x = range(len(label1))
    ax = plt.bar(x, list1,label = '每题得分')

    plt.xlabel('题项',fontsize = 20)
    plt.ylabel('得分',fontsize = 20)
    #plt.ylabel('得分',fontsize = 20)
    mean1 = np.mean(list1).round(2)
   
    plt.axhline(y = mean1,color = 'red', label = 'Mean Value')
    plt.text(np.max(x)/2,mean1,'均值：'+str(mean1),ha = 'center',va = 'bottom',fontsize = 20)

    xlabel1 = [x+1 for x in range(40)]
    plt.xticks([p for p in x],xlabel1,fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.xticks(fontsize = 20)
    plt.legend(fontsize = 20)

    

    for rect in ax:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2, height, str(height), \
            ha="center", va="bottom",fontsize = 20)
    #plt.figure(figsize=(20,15))
    plt.savefig('%s.jpg'%figname1)
    plt.clf()

df = pd.read_excel('./原始数据.xlsx')
#print(df.head(5))

df_describe = df.describe()


#print(np.array(df.iloc[:,4:-1]))
#print(np.array(df.iloc[:,4:-1]).T)


#df.describe().to_excel('./describe.xlsx',encoding='utf-8')
#姓名
label1 = np.array(range(40))


#标题
title1 = np.array(df.iloc[:,[3]]).flatten()
#print(title1)
#print(label1)
#colname = np.array(df.columns)[4:-1]

#print(colname)
#list2 = np.array(df.iloc[:,4:-1])
for i,j,k in zip(np.array(df.iloc[:,4:-1]),np.array(range(12))+1,title1):
    drawfig(df,i,label1,j,k)
```

```python
print(np.array(df.iloc[:,[2]]).flatten())

time = df['所用时间'].str.replace('秒', '').astype(int)
#print(time)
plt.figure(figsize=(12,8))
x = np.linspace(1,12,12)
mean1 = np.mean(time).round(2)
ax = plt.bar(x,time,label = '总耗时')

x_tick = [x+1 for x in range(12)]
plt.xticks([p for p in x],x_tick,fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('被测人员编号',fontsize = 20)
plt.ylabel('做题总耗时（单位：秒）',fontsize = 20)

plt.axhline(np.mean(time),color = 'red',label = '平均耗时')
plt.text(np.max(x)/2,mean1+80,'均值：'+str(mean1),ha = 'center',va = 'bottom',fontsize = 20)

plt.legend(fontsize = 20)

for rect in ax:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2, height, str(height), \
            ha="center", va="bottom",fontsize = 20)
        #plt.figure(figsize=(20,15))
plt.savefig('时间图.jpg')
```

```python
total = df['总分'].astype(int)
#print(time)
plt.figure(figsize=(12,8))
x = np.linspace(1,12,12)
mean1 = np.mean(total).round(2)
ax = plt.bar(x,total,label = '总得分')

x_tick = [x+1 for x in range(12)]
plt.xticks([p for p in x],x_tick,fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('被测人员编号',fontsize = 20)
plt.ylabel('被测人员总得分',fontsize = 20)

plt.axhline(np.mean(total),color = 'red',label = '平均得分')
plt.text(np.max(x)/2,mean1+30,'均值：'+str(mean1),ha = 'center',va = 'bottom',fontsize = 20)

plt.legend(fontsize = 10)

for rect in ax:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2, height, str(height), \
            ha="center", va="bottom",fontsize = 20)
        #plt.figure(figsize=(20,15))
plt.savefig('总分.jpg')
```
